```python
out = torch.zeros(5, 5,).masked_fill(torch.tril(torch.ones(5, 5)) == 0, float('-inf')) # Creates a lower-triangular matrix with 0s and fills the upper triangle with -inf for masking 
out
```

```python
import torch.nn as nn
sample = torch.tensor([10., 10., 10.,])
linear = nn.Linear(3, 3, bias=False) # Applies a linear transformation (no bias) to the input [10, 10, 10] using a randomly initialized 3x3 weight matrix
print(linear(sample))
```

```python
import torch.nn.functional as F
fun_tensor = torch.tensor([1., 2., 3.,])
softmax_output = F.softmax(fun_tensor, dim=0) #exponenties the tensor (2.71^n) and the sums the outputs and then devides the exp. numbers but the sum
print(softmax_output)
```

```python

logits = torch.tensor([[2.0, 0.5, 0.3]])  # prediction for one example
target = torch.tensor([0])  # correct answer is class 0
loss = F.cross_entropy(logits, target)
print(loss)  # smaller if prediction is closer to target
```
